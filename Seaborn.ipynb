# SEABORN CATPLOT

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset/titanic.csv')
df.head()
```

```python
df.Pclass.unique()
```

```python
sb.catplot(x='Sex',data=df,hue='Pclass',kind='count')
plt.show()
```

```python
sb.catplot(x='Pclass',data=df,hue='Sex',kind='count')
plt.show()
```

#this dataset also contains age of children but couldnt depict freq of children

```python
#to find freq of children
def male_female_child(passenger):
    age,gender=passenger
    if(age<16):
        return 'child'
    else:
        return gender
        
```

```python
df['Person']=df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
df.head()
```

```python
sb.catplot(x='Pclass',data=df,hue='Person',kind='count')
plt.show()
```

```python
df['Alone']=df.SibSp
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Alone'].loc[df['Alone']>0]='WithFamily'
df['Alone'].loc[df['Alone']==0]='Alone'
df.head()
```

```python
sb.catplot(x='Alone',data=df,kind='count',palette='rocket')
plt.show()
```
