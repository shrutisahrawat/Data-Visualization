# HOW TO HANDLE MISSING VALUES

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('dataset/weather_data.txt')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()# to check missing recoreds
```

#windspeed,event,temp conatins missing records

```python
df=df.set_index('day')
```

```python
df.head()
```

```python
df[df.temperature.isnull()]
```

```python
df[df.windspeed.isnull()]
```

```python
df[df.event.isnull()]
```

```python
new_df=df.fillna(4)#fills 4 at null value
new_df
```

```python
new_df.isnull().sum()
```

#To make null value will filled value relevant we use mean for temp and windspeed

```python
df
```

```python
df.temperature.mean()
```

```python
df.windspeed.mean()
```

```python
new_df=df.fillna({'temperature':df.temperature.mean(),
                        'windspeed':df.windspeed.mean(),
                         'event':'No Event'})
new_df
```

```python
df
```

```python
new_df=df.bfill()#back value fills
new_df
```

```python
new_df=df.ffill()# forward value fills
new_df
```

#drop null values if data is limited

```python
new_df=df.dropna(axis=0,how='any')#bydefault
new_df#ONLY THE VALUE WITHOUT NULL RECORDS WILL BE LEFT
```

```python
df
```

```python
new_df=df.dropna(axis=1,how='any')#colmn wise
new_df
```

#any in how means here if it will get even a onle null value it will drop whole colmn or row

```python
new_df=df.dropna(axis=0,how='all')#will del  rows with all null values
new_df
```

```python
new_df=df.dropna(axis=0,thresh=3)#thresh=2/greater than 2 null values drop
```

```python
new_df
```

```python
df
```

```python
new_df=df.dropna(subset=['temperature'])#Remove null values of temp
new_df
```

```python
new_df=df.dropna(subset=['temperature','windspeed'])
new_df
```

```python
df=pd.read_csv('dataset/weather_data2.txt')
df.head()
```

```python
new_df= df.replace([-99999,-88888],np.nan)
new_df.head()
```

```python
new_df=df.replace({
    'temperature':[-99999],
    'windspeed':[-99999,-88888],
    'event':'No event'},np.nan)
new_df.head()
```

```python
new_df=new_df.ffill()
new_df.head()
```

```python
df=pd.read_csv('dataset/weather_data3.txt')
df.head()
```

```python
new_df= df.replace('[A-Za-z]',' ',regex=True)
new_df.head()
```

```python
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'},' ',
    regex=True)
new_df.head()
```

```python
new_df.ffill(inplace=True)
new_df.head()
```
