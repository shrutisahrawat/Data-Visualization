### Only those movies starting from Once

```python
movies[movies.title.str.startswith('Once')]
```

# problem is once could be"ONCE",'Once' 

```python
import re
def check(x):
    m=re.search('^once',x,re.I)
    if m:
        return True
    else:
        return False
```

```python
movies[movies.title.apply(check)]# now this will include all once# apply is used in place of loop to call func by checking each index of title and is very fast as comp to loop
```
